Aggregated Transaction

```python
import os
import json
import pandas as pd

# Define the path to the directory containing the aggregated transaction data:
path1 = r"C:\Users\kanis\OneDrive\Desktop\pythonvsworksdir\project\pulse\data\aggregated\transaction\country\india\state/"

#Get the list of directories inside the specified path:
agg_trans_list = os.listdir(path1)# The os.listdir() function returns a list of all the items (directories and files) present in the specified directory.


#Create an empty dictionary to store the data:
columns1 = {'State':[] , 'Year':[] , 'Quarter':[] , 'Transaction_type':[] , 'Transaction_count':[] , 'Transaction_amount':[]} #This dictionary, named columns1, will be used to collect the extracted data.

#Iterate over the directories in agg_trans_list:
for state in agg_trans_list:
    cur_state = path1 + state + "/"        # This loop iterates over each directory in agg_trans_list and concatenates the current state directory's path to the cur_state variable.
    
    if not os.path.isdir(cur_state):
        continue
    
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = cur_state + year + "/"
       
        if not os.path.isdir(cur_year):
            continue
       
        agg_file_list = os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file = cur_year + file 
            
            if not os.path.isfile(cur_file):
                continue
                
                
            #Open and read the JSON file:
            data = open(cur_file,'r')
            A = json.load(data)
            
            #Extract the relevant information from the JSON data and append it to the columns1 dictionary:
            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columns1['Transaction_type'].append(name)
                columns1['Transaction_count'].append(count)
                columns1['Transaction_amount'].append(amount)
                columns1['State'].append(state)
                columns1['Year'].append(year)
                columns1['Quarter'].append(int(file.strip('.json')))
                
df_agg_trans = pd.DataFrame(columns1)              

df_agg_trans["State"] = df_agg_trans["State"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")
df_agg_trans["State"] = df_agg_trans["State"].str.replace("-"," ")
df_agg_trans["State"] = df_agg_trans["State"].str.title()
df_agg_trans["State"] = df_agg_trans["State"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")

df_agg_trans 
```

Aggregated User

```python
path2 = r"C:\Users\kanis\OneDrive\Desktop\pythonvsworksdir\project\pulse\data\aggregated\user\country\india\state/"

agg_user_list = os.listdir(path2)


columns2 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [], 'Percentage': []}

for state in agg_user_list:
    cur_state = os.path.join(path2, state)
    if not os.path.isdir(cur_state):
        continue
    
    agg_year_list = os.listdir(cur_state)
   
    
    for year in agg_year_list:
        cur_year = os.path.join(cur_state, year)
        if not os.path.isdir(cur_year):
          
            continue
        
        agg_file_list = os.listdir(cur_year)
        
        
        for file in agg_file_list:
            if not file.endswith(".json"):
                continue

            cur_file = os.path.join(cur_year, file)
            if not os.path.isfile(cur_file):
               
                continue

            with open(cur_file, 'r') as data:
                B = json.load(data)

            if isinstance(B.get("data", {}).get("usersByDevice"), list):
             for i in B["data"]["usersByDevice"]:
                 columns2["Brands"].append(i["brand"])
                 columns2["Count"].append(i["count"])
                 columns2["Percentage"].append(i["percentage"])
                 columns2["State"].append(state)
                 columns2["Year"].append(year)
                 columns2["Quarter"].append(int(file.replace('.json','')))
            else:
                 pass

df_agg_user = pd.DataFrame(columns2)

df_agg_user["State"] = df_agg_user["State"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")
df_agg_user["State"] = df_agg_user["State"].str.replace("-"," ")
df_agg_user["State"] = df_agg_user["State"].str.title()
df_agg_user["State"] = df_agg_user["State"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")

df_agg_user
```

Aggregated Insurance

```python
path3 = r"C:\Users\kanis\OneDrive\Desktop\pythonvsworksdir\project\pulse\data\aggregated\insurance\country\india\state/"

columns3 = {'State': [], 'Year': [], 'Quarter': [], 'Count': [], 'Amount': []}

for state in os.listdir(path3):
    cur_state = os.path.join(path3, state)
    if not os.path.isdir(cur_state):
        continue

    for year in os.listdir(cur_state):
        cur_year = os.path.join(cur_state, year)
        if not os.path.isdir(cur_year):
            continue

        for file in os.listdir(cur_year):
            if not file.endswith(".json"):
                continue

            cur_file = os.path.join(cur_year, file)
            with open(cur_file, 'r') as f:
                B = json.load(f)

            trans_data = B.get("data", {}).get("transactionData")
            if isinstance(trans_data, list):
                for i in trans_data:
                    if i["name"] == "Insurance":   # ✅ filter only Insurance
                        columns3["State"].append(state)
                        columns3["Year"].append(year)
                        columns3["Quarter"].append(int(file.replace(".json", "")))
                        columns3["Count"].append(i["paymentInstruments"][0]["count"])
                        columns3["Amount"].append(i["paymentInstruments"][0]["amount"])

df_agg_ins = pd.DataFrame(columns3)

df_agg_ins["State"] = df_agg_ins["State"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")
df_agg_ins["State"] = df_agg_ins["State"].str.replace("-"," ")
df_agg_ins["State"] = df_agg_ins["State"].str.title()
df_agg_ins["State"] = df_agg_ins["State"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")

df_agg_ins
```

Creating Table

```python
conn_str = f"mysql+mysqlconnector://{'2UyRh14u8Q2XGo6.root'}:{'9XIoqQ496NnHpCPV'}@{'gateway01.ap-southeast-1.prod.aws.tidbcloud.com'}:{4000}/{'projectphonepe'}"
engine = create_engine(conn_str)


create_tables = [
    """CREATE TABLE IF NOT EXISTS agg_trans (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        Transaction_type VARCHAR(100),
        Transaction_count BIGINT,
        Transaction_amount DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS agg_user (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        Brands VARCHAR(100),
        Count BIGINT,
        Percentage DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS agg_ins (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        Insurance_type VARCHAR(100),
        Count BIGINT,
        Amount DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS map_trans (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        District VARCHAR(100),
        Count BIGINT,
        Amount DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS map_user (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        District VARCHAR(100),
        RegisteredUser BIGINT,
        AppOpens BIGINT
    );""",
    """CREATE TABLE IF NOT EXISTS map_ins (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        District VARCHAR(100),
        Count BIGINT,
        Amount DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS top_trans (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        Pincode VARCHAR(20),
        Count BIGINT,
        Amount DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS top_user (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        Pincode VARCHAR(20),
        RegisteredUser BIGINT
    );""",
    """CREATE TABLE IF NOT EXISTS top_ins (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        Pincode VARCHAR(20),
        Count BIGINT,
        Amount DOUBLE
    );""",
    """CREATE TABLE IF NOT EXISTS map_cor (
        State VARCHAR(100),
        Year INT,
        Quarter INT,
        District VARCHAR(100),
        Lat DECIMAL(20,16),
        Lng DECIMAL(20,16),
        Metric DECIMAL
    );"""
]


with engine.begin() as conn:
    for sql in create_tables:
        conn.execute(text(sql))
print("✅ All tables created successfully!")


dfs = {
    "agg_trans": df_agg_trans,
    "agg_user": df_agg_user,
    "agg_ins": df_agg_ins,
    "map_trans": df_map_trans,
    "map_user": df_map_user,
    "map_ins": df_map_ins,
    "top_trans": df_top_trans,
    "top_user": df_top_user,
    "top_ins": df_top_ins
}

for table, df in dfs.items():
    df.to_sql(table, engine, if_exists="replace", index=False)
    print(f"✅ Inserted {len(df)} rows into {table}")
```
